```python
from kafka import KafkaConsumer # pip install kafka-python

from pprint import PrettyPrinter

import  json
from datetime import datetime

from utils.config import load_config
```

```python
config = load_config("config.yaml")
config
```

```python
# prrpitn
pprint = PrettyPrinter()
```

```python
# Kafka
KAFKA_SERVER = f"{config['kafka']['host']}:{config['kafka']['port']}"
KAFKA_TOPIC_NAME = config['kafka']['topic']
```

```python
consumer = KafkaConsumer(
    KAFKA_TOPIC_NAME,
    group_id=config['kafka']['group-id'],
    bootstrap_servers=KAFKA_SERVER,
    auto_offset_reset=config['kafka']['auto-offset-reset'],
)
```

```python
for msg in consumer:
	pprint.pprint(msg.value)
```
